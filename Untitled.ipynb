```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

min_age = int(input("შეიყვანეთ მინიმალური ასაკი: "))

conn = sqlite3.connect('students.db')
cursor = conn.cursor()

cursor.execute("SELECT Relationship_Status FROM students WHERE Age > ?", (min_age,))
results = cursor.fetchall()

conn.close()

if results:
    df = pd.DataFrame(results, columns=['Relationship_Status'])

    counts = df['Relationship_Status'].value_counts()

    plt.figure(figsize=(10, 6))
    plt.bar(counts.index, counts.values, color='skyblue')
    plt.xlabel('ურთიერთობის სტატუსი')
    plt.ylabel('სტუდენტების რაოდენობა')
    plt.title(f'სტუდენტები ასაკით > {min_age}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print(f"მონაცემები არ მოიძებნა იმ სტუდენტებზე, რომელთა ასაკი მეტია {min_age}-ზე.")
#ეს კოდი გვაჩვენებს ურთიერთობაში მყოფი მოცემულ ასაკზე უფროსი მომხარებლების დიაგრამას.
```

```python
import sqlite3

conn = sqlite3.connect("students.db")
cursor = conn.cursor()

cursor.execute("SELECT Student_ID FROM students ORDER BY Student_ID ASC")
used_ids = [row[0] for row in cursor.fetchall()]

new_id = 1
while new_id in used_ids:
    new_id += 1

print(f"ახალი სტუდენტის ID იქნება: {new_id}")

age = int(input("შეიყვანეთ ასაკი (Age): "))
gender = input("შეიყვანეთ სქესი (Gender): ")
academic_level = input("შეიყვანეთ აკადემიური დონე (Academic_Level): ")
country = input("შეიყვანეთ ქვეყანა (Country): ")
avg_usage = float(input("შეიყვანეთ საშუალო გამოყენების საათები დღეში (Avg_Daily_Usage_Hours): "))
platform = input("შეიყვანეთ ყველაზე ხშირად გამოყენებული პლატფორმა (Most_Used_Platform): ")
affects_perf = input("ახდენს თუ არა გავლენას აკადემიურ წარმატებაზე (Affects_Academic_Performance): ")
sleep_hours = float(input("შეიყვანეთ საშუალო ძილის საათები (Sleep_Hours_Per_Night): "))
mental_health = int(input("შეიყვანეთ მენტალური ჯანმრთელობის ქულა (Mental_Health_Score): "))
relationship = input("შეიყვანეთ ურთიერთობის სტატუსი (Relationship_Status): ")
conflicts = input("კონფლიქტების რაოდენობა სოც. მედიაში (Conflicts_Over_Social_Media): ")
addicted_score = int(input("შეიყვანეთ დამოკიდებულების ქულა (Addicted_Score): "))

cursor.execute("""
    INSERT INTO students (
        Student_ID, Age, Gender, Academic_Level, Country, Avg_Daily_Usage_Hours,
        Most_Used_Platform, Affects_Academic_Performance, Sleep_Hours_Per_Night,
        Mental_Health_Score, Relationship_Status, Conflicts_Over_Social_Media,
        Addicted_Score
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
""", (
    new_id, age, gender, academic_level, country, avg_usage,
    platform, affects_perf, sleep_hours,
    mental_health, relationship, conflicts,
    addicted_score
))

conn.commit()
conn.close()

print(f"ახალი სტუდენტი დამატებულია ID-ით: {new_id}")
#კოდის ეს ნაწილი ამატებს ახალ მონაცემებს, student id  მიენიჭება თავისით, პირველ გამოუყენებლ ადგილზე.
```

```python
import sqlite3

conn = sqlite3.connect("students.db")
cursor = conn.cursor()

delete_id = int(input("შეიყვანეთ წასაშლელი სტუდენტის ID: "))

cursor.execute("DELETE FROM students WHERE Student_ID = ?", (delete_id,))
conn.commit()

conn.close()
#ეს კოდი შეყვანილი ID'ს მიხედვით წაშლის სტუდენტს
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3

conn = sqlite3.connect("students.db")

df = pd.read_sql_query("SELECT * FROM students", conn)

platform_counts = df['Most_Used_Platform'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%')
plt.title("სტუდენტების განაწილება პლატფორმის მიხედვით")
plt.show()
```

```python
age_counts = df['Age'].value_counts().sort_index()
plt.figure(figsize=(8,4))
age_counts.plot(kind='bar', color='skyblue')
plt.title("სტუდენტების რაოდენობა ასაკების მიხედვით")
plt.xlabel("ასაკი")
plt.ylabel("რაოდენობა")
plt.grid(True)
plt.show()
```

```python
avg_age_by_platform = df.groupby('Most_Used_Platform')['Age'].mean()
plt.figure(figsize=(8,4))
avg_age_by_platform.plot(kind='line', marker='o')
plt.title("საშუალო ასაკი პლატფორმების მიხედვით")
plt.xlabel("პლატფორმა")
plt.ylabel("საშუალო ასაკი")
plt.grid(True)
plt.show()

conn.close()
```
